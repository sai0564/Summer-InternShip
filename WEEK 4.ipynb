```python
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [10, 20, 15]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample Data
data = {
    'Age': [20, 21, 22, 23, 24],
    'Marks': [85, 90, 78, 88, 92]
}

df = pd.DataFrame(data)

# Line Plot
plt.plot(df['Age'], df['Marks'])
plt.title("Line Plot")
plt.xlabel("Age")
plt.ylabel("Marks")
plt.show()

# Scatter Plot
plt.scatter(df['Age'], df['Marks'])
plt.title("Scatter Plot")
plt.show()

# Histogram
plt.hist(df['Marks'])
plt.title("Histogram")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Dataset
data = {
    'Age': [20, 21, 22, 23, 24, 25, 26],
    'Marks': [85, 90, 78, 88, 92, 75, 80],
    'StudyHours': [2, 3, 1, 4, 5, 2, 3],
    'SleepHours': [7, 6, 8, 5, 6, 7, 8]
}

df = pd.DataFrame(data)

# Style
sns.set(style="whitegrid")

# 🔹 1. Scatter Plot (Relationship)
plt.figure()
sns.scatterplot(x='StudyHours', y='Marks', data=df)
plt.title("Study Hours vs Marks")
plt.show()

# 🔹 2. Histogram (Distribution)
plt.figure()
sns.histplot(df['Marks'], kde=True)
plt.title("Marks Distribution")
plt.show()

# 🔹 3. Pairplot (All relationships)
sns.pairplot(df)
plt.show()

# 🔹 4. Heatmap (Correlation)
plt.figure()
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```
