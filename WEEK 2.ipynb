```python
from math import factorial

from IPython.terminal.shortcuts.auto_match import braces

#lists
#they are ordered and mutable
#example:
nums = [1,2,3,4,5,6,7,8,9]
nums.append(5)
print(nums)
```

```python
#tuples
#ordered and immutable
#example:
nums = (1,2,3,4)
print(nums)
```

```python
#Dictionaries
#they store data as key value pair
#example:

dict = {"name":"sairam","age":18}
print(dict)
print(dict["name"])
print(dict["age"])
```

```python
#sets
#set is have unique elements and unordered
#example:
sets = {1,2,0,3,2,2,3,4,5,6}
#only prints the unique elements
print(sets)
```

```python
#Funtions
#It is a block of reusable code
#example:
#we need to write a function then we use def
#example:
def wave(name):
    return "hyy "+name
print(wave("sairam!!"))


#recursion function
#calling itself until base case is true
def fact(number):
    if number  ==  1 or number  == 0:
        return number
    else:
        return number*fact(number-1)

print(fact(5))
```

```python
#Lambda function
square = lambda num : num * num
print(square(5))
```

```python
def square(number):
    if number == 0  or number  == 1:
        return number
    else:
        return number * number
print(square(5))
```

```python
def squareList(List):
    return sum(x*x for x in List)
print(squareList([5,6,7]))
```

```python
def filter_odd(list):
    return [x for x in list if x&1 == 1]
print(filter_odd([6,5,9,0,1,2,3,4,5,67,8]))
```

```python
7#CLIENT PROJECT

sample_data = []


while True:
    try:
        num = int(input("Enter number: "))
        sample_data.append(num)
    except:
        break

#removing duplicates

unique_data = set(sample_data)

# clean the data

clean_data = [x for x in unique_data if x  > 0]

clean_data.sort()

print("Original Data",sample_data)
print("Cleaned Data",clean_data)
```
