```python
import pandas as pd
import numpy as np

# Sample dataset (you can replace with CSV file)
data = {
    'Name': ['A', 'B', 'C', 'D', 'E'],
    'Age': [20, 21, np.nan, 22, 20],
    'Marks': [85, 90, 78, np.nan, 88],
    'Department': ['CS', 'IT', 'CS', 'IT', 'CS']
}

df = pd.DataFrame(data)

print("Original Data:")
print(df)

#  Step 1: Handle Missing Values
df_clean = df.dropna()

print("\nAfter Removing Missing Values:")
print(df_clean)

#  Step 2: Fill Missing Values (alternative)
df_fill = df.fillna(df.mean(numeric_only=True))

print("\nAfter Filling Missing Values:")
print(df_fill)

#  Step 3: Calculate Average Marks
avg_marks = df_clean['Marks'].mean()
print("\nAverage Marks:", avg_marks)

#  Step 4: Group By Department
grouped = df_clean.groupby('Department')['Marks'].mean()

print("\nAverage Marks by Department:")
print(grouped)
```
